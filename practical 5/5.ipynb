```r
# Activity 02


movie_types <- c("Comedy", "Action", "Romance", "Drama", "SiFi")
no_of_likes <- c(40, 50, 60, 10, 30)

pie(no_of_likes, 
    labels = movie_types, 
    col = rainbow(length(no_of_likes)), 
    main = "Favorite type of Movie")

pie(no_of_likes, 
    labels = paste0(round(100 * no_of_likes / sum(no_of_likes), 1), "%"), 
    col = rainbow(length(no_of_likes)), 
    main = "Favorite type of Movie")

barplot(no_of_likes, 
        names.arg = movie_types, 
        col = "red", 
        border = "yellow", 
        main = "Favourite type of Movie", 
        xlab = "Movie Type", 
        ylab = "No of Likes")
```

```r
# Activity 04

heart <- read.csv("C:/Users/damsara/Desktop/lab/heart.csv")

boxplot(MaxHR ~ Sex, data = heart,
        main = "Heart Rates",
        xlab = "Sex",
        ylab = "Maximum Heart Rate",
        col = c("purple", "yellow"))

outliers_maxhr <- boxplot.stats(heart$MaxHR)$out
print(outliers_maxhr)

library(moments)
skewness_maxhr <- skewness(heart$MaxHR)
print(skewness_maxhr)

games_file_path <- "C:/Users/damsara/Desktop/lab/games.csv"

if (file.exists(games_file_path)) {
  games <- read.csv(games_file_path)

  print(names(games)) 
  print(head(games))  


  turns_column_name <- "Turns" 

  if (turns_column_name %in% names(games)) { 
      
    boxplot(games[[turns_column_name]] ~ Winner, data = games,  
            main = "Chess game summary",
            xlab = "Winner of the game",
            ylab = "No of turns",
            col = c("green", "blue", "red"))

    outliers_turns <- boxplot.stats(games[[turns_column_name]])$out
    print(outliers_turns)

    skewness_turns <- skewness(games[[turns_column_name]])
    print(skewness_turns)
  } else {
    print(paste("Error: Column '", turns_column_name, "' not found in games.csv", sep=""))
  }


} else {
  print(paste("Error: File not found:", games_file_path))
}


set.seed(123)
mydata1 <- rnorm(100, mean = 50, sd = 10)
boxplot(mydata1, main = "Random Data Boxplot", col = "cyan")


summary(mydata1)

range_val <- range(mydata1)
print(range_val)

skewness_mydata1 <- skewness(mydata1)
print(skewness_mydata1)
```

```r
# Activity 05

data("USArrests")

plot(USArrests$Murder, USArrests$Assault,
     main = "USA arrest rates",
     xlab = "Murders",
     ylab = "Assaults",
     xlim = c(8.0, 14.0),
     ylim = c(150, 300),
     col = "blue",
     pch = 16)

model <- lm(Assault ~ Murder, data = USArrests)
abline(model, col = "red", lwd = 2)
```
